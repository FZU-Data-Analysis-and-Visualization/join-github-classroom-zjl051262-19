```python
%matplotlib inline

import matplotlib.pyplot as plt

plt.rcParams["font.sans-serif"] = ["Microsoft YaHei", "SimHei", "SimSun"]
plt.rcParams["axes.unicode_minus"] = False

x = [
    14, 15, 22, 57, 62, 66, 72, 74, 77, 79, 94, 94,
    100, 106, 109, 129, 137, 171, 172, 183, 193, 198, 260, 365
]
y = [
    0.3, 2.8, 1.1, 0.9, 0.4, 1.0, 0.5, 2.6, 3.1, 3.3, 1.1, 1.7,
    3.3, 1.7, 1.2, 4.1, 6.9, 4.9, 3.6, 12.6, 7.3, 7.8, 10.2, 11.7
]

plt.figure(figsize=(10, 6), facecolor="#f7c48f")
ax = plt.gca()
ax.set_facecolor("#18e1df")

plt.scatter(x, y, s=60, color="red", edgecolors="#75d8d5", linewidths=1.5)

line_x = [12, 370]
line_y = []
for value in line_x:
    y_value = -0.65 + 0.0375 * value
    line_y.append(y_value)

plt.plot(line_x, line_y, color="black", linewidth=2)

plt.xlim(0, 400)
plt.ylim(-2, 14)
plt.xticks([0, 100, 200, 300, 400])
plt.yticks(range(-2, 15, 2))

plt.ylabel("不良贷款", fontsize=18)
plt.title("不良贷款对贷款余额的回归直线", fontsize=22, y=-0.16)
plt.text(390, -1.65, "贷款余额", fontsize=18, ha="right")

for side in ["left", "right", "top", "bottom"]:
    ax.spines[side].set_color("#666666")
    ax.spines[side].set_linewidth(1.8)

ax.tick_params(direction="in", length=6, width=1.5, labelsize=16)

border = plt.Rectangle(
    (0.01, 0.01),
    0.98,
    0.98,
    transform=plt.gcf().transFigure,
    fill=False,
    color="#f39a13",
    linewidth=5
)
plt.gcf().add_artist(border)

plt.tight_layout()
plt.savefig("recreate_loan_plot_code_only.png", dpi=160, facecolor=plt.gcf().get_facecolor())

plt.show()
```
